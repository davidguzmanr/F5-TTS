# Ewe-Igbo-Yoruba (Yoruba New Testament only)

```python
%%bash
python inference-test-split.py \
    --language Yoruba \
    --ckpt_path ckpts/F5TTS_v1_Base_vocos_custom_open-bible-ewe-igbo-yoruba-nt/model_last.pt \
    --vocab_file data/open-bible-ewe-igbo-yoruba-nt_custom/vocab.txt \
    --model_cfg src/f5_tts/configs/F5TTS_v1_Base_Open_Bible_Ewe-Igbo-Yoruba-Nt.yaml \
    --metadata_path data/open-bible-ewe-igbo-yoruba-nt/metadata.csv
```

```python
import os
import pandas as pd
import IPython.display as ipd
from pathlib import Path

ref_index = 0

CKPT_PATH = "ckpts/F5TTS_v1_Base_vocos_custom_open-bible-ewe-igbo-yoruba-nt/model_last.pt"
OUTPUT_DIR = f"synthesis_output/{Path(CKPT_PATH).parent.name}"

test = pd.read_csv(os.path.join(OUTPUT_DIR, "test.csv"))

row = test.iloc[ref_index]
stem = os.path.splitext(row["filename"])[0]
transcript = row["text"]

gen_path = os.path.join(OUTPUT_DIR, "generated", f"{stem}.wav")
gt_path  = os.path.join(OUTPUT_DIR, "ground-truth", f"{stem}.wav")

ipd.display(ipd.Markdown(f"**Transcript:** {transcript}"))
ipd.display(ipd.Markdown("**Generated audio:**"))
ipd.display(ipd.Audio(gen_path))
ipd.display(ipd.Markdown("**Ground-truth audio:**"))
ipd.display(ipd.Audio(gt_path))
```
